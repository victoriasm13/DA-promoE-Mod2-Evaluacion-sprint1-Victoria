<li>Haciendo el anÃ¡lisis exploratorio nos deberÃ­amos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberÃ­an. Esto es debido a que los decimales estÃ¡n establecidos como comas y no con puntos. Cread una funciÃ³n que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.</li>

```python
df.head(3)
```

```python
"""Esta funciÃ³n recoge un string y lo convierte en un nÃºmero float redondeado a dos decimales"""
def aflotar(valor):
    number = valor.replace(",",".")
    return round(float(number),2)
```

```python
columnas = df.columns
columnas
```

```python
str_to_flt = columnas[6:13:1]
str_to_flt
```

```python
for i in str_to_flt:
    df[i] = df[i].apply(aflotar)
```

```python
#Comprobamos que los string se han cambiado a float
df.info()
```

<li>Hay algunas canciones cuyo tÃ­tulo estÃ¡ en mayÃºsulas. Crea una funciÃ³n para que todos los tÃ­tulos estÃ©n en minÃºscula. Haced lo mismo para las columnas de "title" y "album".</li>

```python
def minuscula(sustantivo):
    return sustantivo.lower()
```

```python
col_list = ['album', 'track', 'title']
```

```python
for i in col_list:
    df[i] = df[i].apply(minuscula)
```
